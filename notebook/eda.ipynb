```python
grouped_data=train_data.groupby(["Embarked", "Survived"])["Survived"].count().reset_index(name='Count')

print(grouped_data)
print("\n")

# Plot using seaborn
plt.figure(figsize=(6, 4))  # Adjust the size as needed
sns.barplot(x="Embarked", y="Count", hue="Survived", data=grouped_data)

# Add labels and title
plt.title('Survived Count by Embarked and Survived')
plt.xlabel('Embarked')
plt.ylabel('Count of Survived')

plt.show()
```

## Observations
- People who boarded from "S" had less chance of survival than those who boarded from other 

```python
# Group the data and count 'Sex' for each 'Embarked' and 'Sex'
grouped_data = train_data.groupby(["Embarked", "Sex"])["Sex"].count().reset_index(name="Count")

# Plot using seaborn
plt.figure(figsize=(6, 4))  # Adjust the size as needed
sns.barplot(x="Embarked", y="Count", hue="Sex", data=grouped_data)

# Add labels and title
plt.title('Embarked and Sex Distribution')

plt.show()
```

## Observation
- It is observed hat more males embarked from "S" leading to less survival overall from "S"
- "C" and "Q" had relatively less number of boaders than "S"
